```python
query_string_childcare='https://developers.onemap.sg/privateapi/themesvc/retrieveTheme?queryName=childcare&token=eyJ0eXAiOiJKV1QiLCJhbGciOiJIUzI1NiJ9.eyJzdWIiOjEwMDQ0LCJ1c2VyX2lkIjoxMDA0NCwiZW1haWwiOiJ0YW1teXBvaEBob3RtYWlsLmNvbSIsImZvcmV2ZXIiOmZhbHNlLCJpc3MiOiJodHRwOlwvXC9vbTIuZGZlLm9uZW1hcC5zZ1wvYXBpXC92MlwvdXNlclwvc2Vzc2lvbiIsImlhdCI6MTY3ODk1OTk2NiwiZXhwIjoxNjc5MzkxOTY2LCJuYmYiOjE2Nzg5NTk5NjYsImp0aSI6ImZiZGVmMmY3ZjYxMTE5NmQ0NGFhYTQ3M2JlMjc3Y2ZjIn0.yrvyqexznNvkb9pHDGajkQsafIbzGYZpbz2Gdfk3umY'
resp_childcare = requests.get(query_string_childcare)
childcare=json.loads(resp_childcare.content)
childcare['SrchResults'][1]
```

```python
def get_data(feature, amenity_type):
    
    name = []
    lat_lng = []
    
    for i in range(1,len(feature['SrchResults'])):
        name.append(feature['SrchResults'][i]['NAME'])
        lat_lng.append(feature['SrchResults'][i]['LatLng'])
    
    df = pd.DataFrame(list(zip(name, lat_lng)),
                        columns =['Name','lat_lng'])
    df[['latitude','longtitude']] = df['lat_lng'].str.split(',',expand=True)
    df = df.drop('lat_lng', axis=1)
    df.insert(0, 'Feature', amenity_type)
    
    return (df)
```

```python
preschool_df = get_data(preschools, 'pre-school')
kindergartens_df = get_data(kindergartens, 'kindergarten')
childcare_df = get_data(childcare, 'childcare centers')

child_education = pd.concat([preschool_df, kindergartens_df, childcare_df], ignore_index=True)
child_education.head()
```

```python
#drop duplicates
child_education.drop_duplicates(subset= ['Name', 'latitude', 'longtitude'],keep= 'first', inplace = True)
child_education = child_education.reset_index(drop=True)
child_education
```

```python
child_education.to_csv('child_education_coordinates.csv')
```
